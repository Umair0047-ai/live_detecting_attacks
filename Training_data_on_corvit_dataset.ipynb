```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_csv("Corvit_Implemented_tcp_flood.csv")
df2 = pd.read_csv('Corvit_Implemented_udp_flood.csv',error_bad_lines=False)
df3 = pd.read_csv('Corvit_normal_data.csv')
df4 = pd.read_csv('corvit_rtsp.csv')
```

```python
df = df.fillna(0)
df2 = df2.fillna(0)
df3 = df3.fillna(0)
df4 = df4.fillna(0)
```

```python
df = df.drop(['frame.len','ip.src','ip.dst','ipv6.src','ipv6.dst','eth.src','eth.dst','ip.id','tcp.flags','ipv6.dst','ip.ttl','icmp.length','http.request.method','http.response.code','http.content_length'],axis = 1)
df = pd.get_dummies(df)
df["attack_type"] = "TCP_flood"
df
```

```python
df2 = df2.drop(['frame.len','ip.src','ip.dst','ipv6.src','ipv6.dst','eth.src','eth.dst','ip.id','tcp.flags','ipv6.dst','ip.ttl','icmp.length','http.request.method','http.response.code','http.content_length'],axis = 1)
df2 = pd.get_dummies(df2)
df2["attack_type"] = "Udp_flood"
df2
```

```python
df3 = df3.drop(['frame.len','ip.src','ip.dst','ipv6.src','ipv6.dst','eth.src','eth.dst','ip.id','tcp.flags','ipv6.dst','ip.ttl','icmp.length','http.request.method','http.response.code','http.content_length'],axis = 1)
df3 = pd.get_dummies(df3)
df3["attack_type"] = "Normal_traffic"
df3
```

```python
df4 = df4.drop(['ip.src','ip.dst','ipv6.src','ipv6.dst','eth.src','eth.dst','ip.id','tcp.flags','ipv6.dst','ip.ttl','icmp.length','http.request.method','http.response.code','http.content_length'],axis = 1)
df4 = pd.get_dummies(df4)
df4["attack_type"] = "Brute_force"
df4
```

```python
frames = [df,df2,df3,df4]

### concatenating data
result = pd.concat(frames)

##display
result
```

```python
y = result['attack_type']
y
```

```python
### dropping type because it will be our label that we want to predict 
X = result.drop("attack_type",axis=1)

### display
X
```

```python
#### implementing Random Forest
## using Train TESt split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
import time
```

```python
start_time = time.time()
rfc = RandomForestClassifier(n_estimators=60,max_features="auto",random_state=101) 
rfc.fit(X_train,y_train)
end_time = time.time() 
```

```python
train_time = end_time - start_time
print("Training time :",train_time,"seconds") 
```

```python
start_time = time.time()
preds = rfc.predict(X_test)
end_time = time.time()
```

```python
test_time = end_time - start_time
print("Test time :",test_time,"seconds")
```

```python
from sklearn.metrics import classification_report,confusion_matrix,plot_confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,preds))
model_accuracy = accuracy_score(y_test,preds)
print("Model Accuracy:", model_accuracy)
```
